```python
# replacing spaces with '\n' to make long strings comes in second line
Data_all.iloc[:,0]=Data_all.iloc[:,0].str.replace(' ','\n',1)
```

```python
#Remove '+' with '\n+'
Data_all.columns=Data_all.columns.str.replace('+','\n+')
Data_all.columns=Data_all.columns.str.replace(' ','',1)
Data_all.columns=Data_all.columns.str.replace('(','\n(')
```

```python
Data_all=Data_all.set_index(Data_all.iloc[:,0])
```

```python
Data_all=Data_all.drop(Data_all.columns[[0]], axis=1)
```

```python
s_count=pd.read_csv("../Tables_Created/bac_vs_antibiotic_s_count2.csv")
all_count=pd.read_csv("../Tables_Created/bac_vs_antibiotic_all_count2.csv")
```

```python
print(s_count.shape)
print(all_count.shape)
```

```python
#s_count=s_count.drop(s_count.columns[[0]], axis=1)  # df.columns is zero-based pd.Index 
s_count=s_count.set_index(s_count.iloc[:,0])
s_count=s_count.drop(s_count.columns[[0]], axis=1)  # df.columns is zero-based pd.Index 
s_count.index.name = 'Bacteria'
```

```python

#all_count=all_count.drop(all_count.columns[[0]], axis=1)  # df.columns is zero-based pd.Index 
all_count=all_count.set_index(all_count.iloc[:,0])
all_count=all_count.drop(all_count.columns[[0]], axis=1)  # df.columns is zero-based pd.Index 
all_count.index.name = 'Bacteria'
```

```python
labels=[]
flatten_data_all=[]
import math
for value1,value2,value3 in zip(s_count.values.flatten(),all_count.values.flatten(),Data_all.values.flatten()):
    #print(value1,value2,round(value3,3))
    #if value1==0 and value2==0:
    if math.isnan(value3):
        labels.append("Nan")
    else:
        if len(str(value2))>3:
            labels.append("{0} /\n{1}".format(value1,value2,round(value3,2)))
        else:
            labels.append("{0} / {1}".format(value1,value2,round(value3,2)))

labels=np.array(labels)
labels=labels.reshape(Data_all.shape)
```
