## Discharge Summary

#### ALL DISCHARGE SUMMARIES

```python
dissums = note[note['CATEGORY']=='Discharge summary'].reset_index(drop=True)
```

```python
dissums.DESCRIPTION.value_counts()
```

```python
import re
def get_list_of_entries(text):
    text = text.replace("\n", " ")
    text = re.sub(' +', ' ', text)
    matches = re.findall(r'(^[A-Z][a-zA-Z\s]*)\:', text)
    if len(matches)>0:
        text = re.sub(matches[0],'', text)
    matches = matches + re.findall(r'\s([A-Z][a-zA-Z\s]*)\:', text)
    return matches
```

```python
dissums['TOPICS'] = dissums['TEXT'].apply(lambda x: get_list_of_entries(x))
```

```python
all_topics = []
for i in range(dissums.shape[0]):
    all_topics.append(dissums.iloc[i].TOPICS)
unique_topics = set().union(*all_topics)
print(len(unique_topics))
```

```python
common_topics = set(all_topics[0]).intersection(*all_topics)
print(common_topics)
```

```python
list(unique_topics)
```

#### REPORTS IN DISCHARGE SUMMARIES

```python
reports = dissums[dissums['DESCRIPTION']=='Report'].reset_index(drop=True)
```

```python
reports.shape
```

```python
reports['TOPICS'] = reports['TEXT'].apply(lambda x: get_list_of_entries(x))
```

```python
all_topics_report = []
for i in range(reports.shape[0]):
    all_topics_report.append(reports.iloc[i].TOPICS)
unique_topics_report = set().union(*all_topics_report)
print(len(unique_topics_report))
```

```python
common_topics_report = set(all_topics_report[0]).intersection(*all_topics_report)
print(common_topics_report)
```

```python
list(unique_topics_report)
```

```python
all_topics_report_list = []
for i in range(reports.shape[0]):
    all_topics_report_list = all_topics_report_list + reports.iloc[i].TOPICS
len(all_topics_report_list)
```

```python
from collections import Counter
count_cases = Counter(all_topics_report_list)
```

```python
count_cases.most_common(100)
```
